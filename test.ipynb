```python
faturamento = vendas_df[['ID Loja','Valor Final']].groupby('ID Loja').sum()
quantidade = vendas_df[['ID Loja','Quantidade']].groupby('ID Loja').sum()
```

```python
ticket_medio = (faturamento['Valor Final'] / quantidade['Quantidade']).to_frame()

#ALTERAR COLUNA
ticket_medio = ticket_medio.rename(columns={0: 'Ticket Medio'})
#FORMAT
ticket_medio_formatado =  ticket_medio.style.format({'Ticket Medio': 'R${:,.2F}'})

ticket_medio_xlsx = ticket_medio_formatado.to_excel('ticket_medio.xlsx' )

ticket_medio_csv = ticket_medio.to_csv('ticke_medio.csv')
```
